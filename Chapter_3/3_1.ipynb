```python
# ファイルを読み込む。
file = open(
'/content/drive/MyDrive/test.txt')
# ファイルのテキストを取り出す。
text = file.read()
text
```

```python
# 単語ごとにリストにする。
words = text.split()
words
```

```python
# string モジュールをインポートする。
import string
# 句読点などの記号類を指定する。
string.punctuation
```

```python
'''abc においてa をA、c をC に置き換え、b とe を削除
する。'''
'abcde'.translate(
str.maketrans('ac', 'AC', 'be'))
```

```python
''' 単語リストのそれぞれの語の中で、句読点などの記号が
あるものはその記号を削除する。'''
words = [
word.translate(str.maketrans(
'', '', string.punctuation))
for word in words
]
words
```

```python
# 単語リスト中のすべての文字を小文字にする。
words = [word.lower() for word in words]
words
```

```python
# Natural Language Toolkit をインポートする。
import nltk
# WordNetLemmatizer をインポートする。
from nltk.stem import WordNetLemmatizer
# 必要なデータをダウンロードする。
nltk.download('omw-1.4')
nltk.download('wordnet')
# 見出し語化（単語を原型に戻す）処理をおこなう。
words = [WordNetLemmatizer().lemmatize(word)
for word in words]
words
```

```python
# ストップワードのリストをダウンロードする。
nltk.download('stopwords')
# 英語のストップワードのリストを呼び出す。
nltk.corpus.stopwords.words('english')
```

```python
''' ab, bc, cd のうち、ab, cd の要素でないものだけをプ
リントする。'''
for item in ['ab', 'bc', 'cd']:
  if item not in ['ab', 'cd']:
    print(item)
```

```python
words = [word for word in words
if word not in
nltk.corpus.stopwords.words\
('english')]
words
```

```python
# ファイルを読み込む。
file = open(
'/content/drive/MyDrive/my_vocab.txt')
# ファイルのテキストを取り出す。
my_vocab = file.read()
# 既知語のリストを作る。
my_vocab = my_vocab.split()
my_vocab
```

```python
# 単語リストから既知の単語を除去する。
words = [word for word in words
if word not in my_vocab]
words
```

```python
# 単語の出現回数を数える。
from collections import Counter
# Counter を使って、単語の出現回数を計算する。
freq = Counter(words)
# 出現回数を表示する。
freq
```

```python
'''3 回以上出現し、文字数が7 以上である語のみ抽出し、
target という変数に格納する。'''
target = [word for word,
frequency in freq.items()
if frequency >= 3 and \
len(word) >= 7]
target
```

```python
# pandas をインポートする。
import pandas as pd
'''read_table を使用して、Drive にアップロードした英
和辞書を読み込む。'''
df = pd.read_table(
'/content/drive/My Drive/'\
'ejdict-hand-utf8.txt',
header=None)
'''0 番目の列と1 番目の列をそれぞれセットにして、辞書
形式のデータにする。'''
dictionary = dict(zip(df[0], df[1]))
dictionary
```

```python
''' 対象の単語が辞書に登録されている場合、単語とその意
味をペアで出力する。'''
for t in target:
  if t in dictionary:
    print(t, dictionary[t])
```
